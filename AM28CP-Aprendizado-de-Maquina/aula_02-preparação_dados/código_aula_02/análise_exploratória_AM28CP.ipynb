## Visualização de dados

```python
_, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 10))

for att, axis in enumerate(axes.flat):
    axis.hist(np.array(X[X.columns[att]]))
    axis.set_xlabel(X.columns[att])
```

```python
labels = np.array(iris['class'].unique())

colors = ['red', 'green', 'blue']
```

```python
_, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 10))

for att, axis in enumerate(axes.flat):
    for label, color in zip(range(len(labels)), colors):
        axis.hist(np.array(X[X.columns[att]][iris['class'] == labels[label]]),
                  label = labels[label], color = color)
    
    axis.set_xlabel(X.columns[att])  
    axis.legend(loc='upper right')

```

```python
_, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (7, 7))

for att, axis in enumerate(axes.flat):
    axis.boxplot(np.array(X[X.columns[att]]))
    axis.set_xlabel(X.columns[att])
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 10))

for att, axis in enumerate(axes.flat):
    for label in range(len(labels)):
        axis.boxplot(np.array(X[X.columns[att]][iris['class'] == labels[label]]), positions = [label])
    
    axis.set_xlabel(X.columns[att])
    axis.set_xticklabels(labels, color = 'blue')
```

```python
_, axes = plt.subplots(nrows = 4, ncols = 4, figsize = (20, 20))

for i in range(len(X.columns)):
    for j in range(len(X.columns)):
        scat = axes[i][j].scatter(X[X.columns[i]], X[X.columns[j]], c = target)
        axes[i][j].set(xlabel = X.columns[i], ylabel = X.columns[j])
        axes[i][j].legend(scat.legend_elements()[0], labels, loc = "upper right", title = "Classes")
    
```
