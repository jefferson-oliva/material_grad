```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
```

## PCA from scratch

```python
data = load_iris()
X = data.data   # atributos
y = data.target # classes
```

#### Padronização: média 0, desvio padrão 1

```python
X_std = (X - X.mean(axis = 0)) / X.std(axis = 0)
```

#### Cálculo da matriz de covariância

```python
cov_matrix = np.cov(X_std.T)
```

#### Cálculo de autovalores e autovetores

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

#### Ordenação dos autovetores e autovalores

```python
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]
```

#### Seleção dos k componentes principais

```python
k = 2
pc = eigenvectors[:, : k]
```

#### Projeção dos dados e novo espaço

```python
X_pca = X.dot(pc)
np.shape(X_pca)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = y, cmap = 'viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA - Iris Dataset')
```

## PCA via scikit-learn

```python
X = data.data
y = data.target
```

```python
pca = PCA(n_components = 2)
```

```python
X_pca = pca.fit_transform(X)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = y)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA - Iris Dataset')
plt.show()
```
