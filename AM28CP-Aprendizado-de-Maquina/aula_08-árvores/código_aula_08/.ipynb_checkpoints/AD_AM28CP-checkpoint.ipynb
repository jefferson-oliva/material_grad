## Exemplo de treinamento de uma árvore de decisão

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

#### Abertura da base de dados

```python
bc = datasets.load_breast_cancer()
x = bc.data
y = bc.target
```

#### Divisão do conjunto de dados entre treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, shuffle = True, random_state = 1, stratify = y)
```

```python
print('Labels counts in y:', np.bincount(y))
print('Labels counts in y_train:', np.bincount(y_train))
print('Labels counts in y_test:', np.bincount(y_test))
```

#### Normalização dos dados

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_std = scaler.transform(X_train)

scaler.fit(X_test)
X_test_std = scaler.transform(X_test)
```

#### Treinamento de uma árvore de decisão

```python
model = DecisionTreeClassifier(criterion = 'gini', max_depth = 3)

model.fit(X_train_std, y_train)
```

#### Predição e avaliação

```python
y_pred = model.predict(X_test_std)
```

```python
accuracy_score(y_test, y_pred)
```

#### Plotar a árvore resultante

```python
plt.figure(figsize=(30, 45))

plot_tree(model,
          filled=True,
          rounded=True,
          class_names= bc.target_names,
          feature_names=bc.feature_names)

plt.show()
```
