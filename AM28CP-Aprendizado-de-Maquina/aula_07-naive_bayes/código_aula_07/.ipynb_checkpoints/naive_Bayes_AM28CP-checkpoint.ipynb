```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import make_classification
```

## Exemplo para teorema de Bayes

```python
P_doente = 0.01
P_nao_doente = 1 - P_doente
P_positivo = 0.06

#P(positivo | doente)
P_positivo_dado_doente = 0.99

#P(positivo | não doente)
P_positivo_dado_nao_doente = 0.05
```

```python
# Cálculo da posteriori: P(doente | positivo)
P_doente_dado_positivo = (P_positivo_dado_doente * P_doente) / P_positivo
```

```python
labels = ['Doente', 'Não doente']
prior = [P_doente, P_nao_doente]
posterior = [P_doente_dado_positivo, 1 - P_doente_dado_positivo]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 6))
bars1 = ax.bar(x - width / 2, prior, width, label = 'A priori')
bars2 = ax.bar(x + width / 2, posterior, width, label = 'A posteriori')

ax.set_ylabel('Probabilidade')
ax.set_title('Probabilidade de estar doente antes e depois do teste positivo')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.set_ylim(0, 1)

# Adicionando rótulos acima das barras
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}', xy = (bar.get_x() + bar.get_width() / 2, height),
                xytext = (0, 3), textcoords = "offset points",
                ha = 'center', va = 'bottom')

plt.tight_layout()
plt.show()
```

## Construindo um classificador naïve Bayes

#### Carregar os dados

```python
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

#### Dividir em treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

#### Treinar um classificador

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

#### Predição e avaliação de desempenho

```python
y_pred = model.predict(X_test)
```

```python
print("Acurácia:", accuracy_score(y_test, y_pred))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred, target_names=target_names))
```

## Exemplo 2

```python
# Gerar dados artificiais
X, y = make_classification(n_samples = 400, n_features = 2, n_redundant = 0, 
                           n_informative = 2, n_clusters_per_class = 1, 
                           class_sep = 1.5, random_state = 42)
```

```python
# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Treinar o classificador
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Acurácia:", accuracy_score(y_test, y_pred))
```

```python
# Visualizar as regiões de decisão
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 400), np.linspace(y_min, y_max, 400))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
plt.title("Naive Bayes com dados 2D simulados")
plt.xlabel("Atributo 1")
plt.ylabel("Atributo 2")
plt.grid(True)
plt.show()
```
