# Morfologia matemática

### 1. Erosão

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('j.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.axis('off')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(img, kernel, iterations = 1)
```

```python
plt.axis('off')
plt.imshow(cv2.cvtColor(erosion, cv2.COLOR_BGR2RGB))
```

### 2. Dilatação

```python
dilation = cv2.dilate(img, kernel, iterations = 1)
```

```python
plt.axis('off')
plt.imshow(cv2.cvtColor(dilation, cv2.COLOR_BGR2RGB))
```

### 3. Abertura

```python
img_a = cv2.imread('j_ruido.png', cv2.IMREAD_GRAYSCALE)
plt.title("Imagem original")
plt.axis('off')
plt.imshow(cv2.cvtColor(img_a, cv2.COLOR_BGR2RGB))

opening = cv2.morphologyEx(img_a, cv2.MORPH_OPEN, kernel)
```

```python
plt.title("Imagem aplicada abertura")
plt.axis('off')
plt.imshow(cv2.cvtColor(opening, cv2.COLOR_BGR2RGB))
```

### 4. Fechamento


```python
img_a = cv2.imread('j_orificios.png', cv2.IMREAD_GRAYSCALE)
plt.title("Imagem original")
plt.axis('off')
plt.imshow(cv2.cvtColor(img_a, cv2.COLOR_BGR2RGB))
```

```python
closing = cv2.morphologyEx(img_a, cv2.MORPH_CLOSE, kernel)

plt.title("Imagem aplicado fechamento")
plt.axis('off')
plt.imshow(cv2.cvtColor(closing, cv2.COLOR_BGR2RGB))
```

### 5. Gradiente morfológico: difença entre dilatação e erosão de uma imagem.

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

plt.title("Imagem aplicado gradiente")
plt.axis('off')
plt.imshow(cv2.cvtColor(gradient, cv2.COLOR_BGR2RGB))
```

### 6. Top hat: diferença entre a imagem de entrada e a abertura da imagem

```python
kernel = np.ones((9, 9),np.uint8)

tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

plt.title("Imagem aplicado top hat")
plt.axis('off')
plt.imshow(cv2.cvtColor(tophat, cv2.COLOR_BGR2RGB))
```

### 7. Black hat: diferença entre o fechamento da imagem de entrada e a imagem de entrada

```python
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT, kernel)

plt.title("Imagem aplicado Black hat")
plt.axis('off')
plt.imshow(cv2.cvtColor(blackhat, cv2.COLOR_BGR2RGB))
```

### 8. Elemento estruturante

```python
# Kernel retangular
cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
```

```python
# Kernel elíptico
cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
```

```python
# Kernel 'encruzilhada'
cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))
```
