```python
import cv2
import os
import numpy as np
import pathlib
import matplotlib.pyplot as plt
```

## Template Matching

```python
img1 = cv2.imread('marioworld.png')  
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray1, cv2.COLOR_BGR2RGB))
```

```python
## carregando o template
template = cv2.cvtColor(cv2.imread('templatemario.png'), cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(template, cv2.COLOR_BGR2RGB))
```

```python
h, w = template.shape

print(w)
print(h)
```

```python
res = cv2.matchTemplate(gray1, template, cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

```

```python
gr = gray1.copy()

# A definição da variável abaixo depende do coeficiente de normalização escolhido
# Por exemplo, caso a escolha tenha sido TM_SQDIFF ou cv.TM_SQDIFF_NORMED, então
# top_left = min_loc
top_left = max_loc

bottom_right = (top_left[0] + w, top_left[1] + h)

cv2.rectangle(gr, top_left, bottom_right, 255, 2)

plt.imshow(gr, cmap = 'gray')
```
